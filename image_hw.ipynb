```python
import numpy as np
from PIL import Image
import requests
import matplotlib.pyplot as plt
```

```python
url = 'https://people.com/thmb/sQSl8jHALt3PIvnpVV2uUWxB1ew=/1500x0/filters:no_upscale():max_bytes(150000):strip_icc():focal(599x0:601x2)/BTS_JOIN_LOUIS_VUITTON_AS_HOUSE_AMBASSADORS-a525d86a4d4745dd8ff413cc723c6c7e.jpg'
photo = np.asarray(Image.open(requests.get(url, stream=True).raw))
```

```python
print(photo)
```

```python
def red_filter(photo):
  red_only = np.copy(photo)
  red_only[:, :, 1:] = 0
  return red_only
```

```python
def green_filter(photo):
  green_only = np.copy(photo)
  green_only[:, :, [0, 2]] = 0
  return green_only
```

```python
def blue_filter(photo):
  blue_only = np.copy(photo)
  blue_only[:, :, :2] = 0
  return blue_only
```

```python
plt.imshow(photo)
plt.title("Original Image")
plt.show()

red_scale = red_filter(photo)
plt.imshow(red_scale)
plt.title("Red Scale Image")
plt.show()

green_scale = green_filter(photo)
plt.imshow(green_scale)
plt.title("Green Scale Image")
plt.show()

blue_scale = blue_filter(photo)
plt.imshow(blue_scale)
plt.title("Blue Scale Image")
plt.show()

rg_scale = np.copy(photo)
rg_scale[:, :, [2]] = 0
plt.imshow(rg_scale)
plt.title("Red Green Scale Image")
plt.show()

rb_scale = np.copy(photo)
rb_scale[:, :, [1]] = 0
plt.imshow(rb_scale)
plt.title("Red Blue Scale Image")
plt.show()

gb_scale = np.copy(photo)
gb_scale[:, :, [0]] = 0
plt.imshow(gb_scale)
plt.title("Green Blue Scale Image")
plt.show()

check = red_scale + green_scale + blue_scale
plt.imshow(check)
plt.title("Recreated Image")
plt.show()
if np.array_equal(check, photo):
  print("Photos are the same")
else:
  print("Photos are NOT the same")
```
